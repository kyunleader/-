# 실습 2 날씨데이터

```python
#데이터 로드
data_w = pd.read_csv('data_weather.csv')
```

```python
data_w
```

```python
data_w.columns
```

```python
# 분석에 사용할 x데이터
data_x = data_w.loc[:,['평균기온(°C)','평균최고기온(°C)',
       '평균최저기온(°C)', '평균일교차(°C)', '평균상대습도(%)', '월합강수량(00~24h만)(mm)',
       'SO2(ppm)', 'NO2(ppm)', 'CO(ppm)', 'PM-10(ug/m3)']]
```

```python
# 변수간 상관관계 분석
data_corr = data_x.corr()  # 간단히 상관관계 구하기
data_corr
```

```python
# 변수간 상관관계 분석
import seaborn as sns

# 기본적인 상관관계 그래프
sns.clustermap(data_corr,
               annot=True,  # 실제 값 화면에 나타내기
               cmap='RdYlBu_r',  # Red, Yellow, Blue 로 색상 표시
               vmin=-1, vmax=1  # 컬러차트 -1 ~ 1 범위로 표시
               )
```

```python
# 분석에 사용할 x데이터
data_x = data_w.loc[:,['평균기온(°C)', '평균일교차(°C)', '평균상대습도(%)', '월합강수량(00~24h만)(mm)',
       'SO2(ppm)', 'NO2(ppm)', 'CO(ppm)', 'PM-10(ug/m3)']]
```

```python

# 주성분 분석 해보기

# 정규화 시키기
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(data_x)
```

```python
# 주성분 분석 (PCA)
from sklearn.decomposition import PCA

# 그래프로 표현해보기
pca = PCA()
pca.fit(x)

exp_var_cumul = np.cumsum(pca.explained_variance_ratio_)
exp_var_cumul2 = pd.DataFrame(exp_var_cumul, index=range(1, exp_var_cumul.shape[0] + 1))
# print(exp_var_cumul)
# print(exp_var_cumul2)
pd.DataFrame(pca.components_, columns=[data_x.columns], index=[str(i+1) + '번 주성분' for i in range(8)]).round(2)
```

```python
print(pca.explained_variance_)

print(exp_var_cumul2)
```

```python

plt.ylim(0, 1.1)
sns.pointplot(data=exp_var_cumul2.T, palette='Blues_d')

```

```python
# 주성분 데이터 셋 만들기
pca = PCA(n_components=2)  # PCA 객체 생성 (2개)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data=principalComponents,
                           columns=['principalComponents1', 'principalComponents2'])  # 2개의 주성분이 나타내는 값
print(principalComponents)
```

```python

# 2개의 인자들이 얼마만큼 변수를 설명 가능한지
print(pca.explained_variance_ratio_)

# 누적 값
print(sum(pca.explained_variance_ratio_))
```

```python
# 2차원으로 축소된 데이터
data_2d = pd.concat([principalDf, data_w['발생건수(건)']], axis=1)
```

```python
data_2d
```

```python
plt.scatter(x = data_2d['principalComponents1'], y = data_2d['principalComponents2'], c = data_2d['발생건수(건)'])
plt.colorbar()
```
